## Generating stable laws using deterministic dynamical *systems*

It was shown in [Gottwald-Melbourne (2021)](https://epubs.siam.org/doi/10.1137/20M1333183) that $\alpha$-stable laws, $\alpha \in (1,2)$, can be generated using a deterministic dynamical system whose states $x_n$ are obtained by iterating a map $T$: 
$$x_{n+1} = T(x_n),$$
with any source of randomness coming solely from the initial condition $x_0 \in [0,1]$.

As in [Gottwald-Melbourne (2021)](https://epubs.siam.org/doi/10.1137/20M1333183), we consider observables $v$ of maps introduced in the study of intermittency by Pomeau-Manneville. In particular, the family of maps $T: [0,1] \to [0,1]$ considered by Thaler:

$$T(x) = (x^{1-\gamma} + (1+x)^{1-\gamma} - 1)^{1/(1-\gamma)} \mod 1,$$

with $\gamma \in [0,1) \cup (1,\infty)$ a real parameter. 


Let $x^* \in (0,1)$ be the unique solution to the equation:

$$(x^*)^{1-\gamma} + (1+x^*)^{1-\gamma} = 2.$$
There are two branches (increasing) defined on the intervals $[0,x^*]$, $[x^*, 1]$ (see Fig. 1 in [Gottwald-Melbourne (2021)](https://epubs.siam.org/doi/10.1137/20M1333183)).

Unlike other intermittent maps considered in the literature, the Thaler map allows for analytic expressions for the map and the invariant density. For $\gamma \in [0,1)$, there exists a unique invariant probability density $\tilde{h} = \frac{1-\gamma}{2^{1-\gamma}} h$, where

$$h(x) = x^{-\gamma} + (1+x)^{-\gamma}.$$

The relevant case for us is when $\gamma \in (0,1)$, where the Thaler map is nonuniformly expanding with a neutral fixed point at $x=0$ and correlations decay algebraically with rate $1/n^{1/\gamma - 1}$ (which is sharp). The slow down in the decay of correlation as $\gamma$ increases is caused by the trajectory spending prolonged times near the neutral fixed point $x=0$ (laminar dynamics near the fixed point). Note that the density $h$ defines a finite measure only for $\gamma < 1$. 

Let $v: [0,1] \to \mathbb{R}$ be a Holder observable and suppose that $v$ has mean zero with respect to the invariant probability measure $d\mu = h dx$.  We further restrict to the case when $\gamma \in (1/2, 1)$, where the correlations are not summable and the CLT breaks fown for observables with $v(0) \neq 0$ that ''see'' the neutral fixed point at $x=0$. Heuristically, the Birkhoff sum $v_n := \sum_{k=0}^{n-1} v \circ T^k$ experiences ballistic behavior with almost linear behavior in the laminar region near $x=0$ and the small jumps of size $v(0)$ accumulate into a single large jump incompatible with the CLT. Instead, the CLT is replaced by a one-sided stable limit law 
$$n^{-\gamma} v_n \to_d X_{\alpha, \eta, \beta}$$ with $\alpha = 1/\gamma$ and $\beta = sign(v(0))$, where $X_{\alpha, \eta, \beta}$ denote stable laws whose characteristic function is given by: 
$$\mathbb{E}[e^{it X_{\alpha, \eta, \beta}}] = \exp(-\eta^\alpha |t|^\alpha (1-i\beta sign(t) \tan(\alpha \pi/2))),$$
where $\alpha \in (0,1) \cup (1,2)$, $\eta > 0$ and $\beta \in [-1,1]$. 

Let $T$ be the Thaler map with $\gamma = 1/\alpha \in (1/2, 1)$, 
$$d_\alpha = \frac{\alpha^\alpha (1-\gamma) \Gamma(1-\alpha) \cos(\alpha \pi/2)}{2^{1-\gamma}-1},$$
and $\delta_0, \delta_1, \dots$ are independent copies of the random variable $\delta$ where $\mathbb{P}(\delta = \pm 1) = \frac{1}{2}(1 \pm \beta)$. 
Define a sequence of observables $v^{(n)} = \chi^{(n)} v \circ T^n$, where $v: [0,1] \to \mathbb{R}$ is the mean zero observable given by: 
$$v(x) = \eta d_\alpha^{-\gamma} (1-2^{\gamma-1})^{-\gamma} \tilde{v}(x),$$

$$\tilde{v}(x) = \begin{cases}
  1 & \text{if $x \leq x^*$}, \\
  (1-2^{1-\gamma})^{-1} & \text{if $x > x^*$},
\end{cases} $$

and

$$\chi^{(n)} = \chi_{n-1} \cdots \chi_0 \in \{\pm 1 \}, $$

$$\chi_k = \begin{cases}
  1 & \text{if $T^k x \leq x^*$}, \\
  \delta_k & \text{if $T^k x > x^*$}.
\end{cases} $$ 

In particular, the random variable $\chi^{(n)}$ gets updated only when the trajectory visits $(x^{*}, 1]$ and is unchanged during the laminar phase in $[0, x^*]$. 

**Theorem** (Theorem 4.1 in [Gottwald-Melbourne (2021)](https://epubs.siam.org/doi/10.1137/20M1333183)). Let $a: \mathbb{R} \to \mathbb{R}$ be $C^{1+\delta}$ and $b: \mathbb{R} \to \mathbb{R}$ be $C^{\alpha + \delta}$ for some $\delta > 0$. Fix $\xi \in \mathbb{R}$ and consider the fast-slow map defined by:

$$z_{n+1}^\epsilon = z_n^\epsilon + \epsilon a(z_n^\epsilon) + \epsilon^\gamma b(z_n^\epsilon) v^{(n)}, \ z_0^\epsilon = \xi,$$

where $v^{(n)}: [0,1] \to \mathbb{R}$ is defined as above. 

Set $\hat{z}_\epsilon(t) = z^\epsilon_{\lfloor t/\epsilon \rfloor}$. Then $\hat{z}_\epsilon$ converges weakly to $Z$ in $D([0,\infty))$ on the probability space $([0,1], \mu)$ as $\epsilon \to 0$, where $Z$ is the solution to the Marcus SDE (satisfying the standard laws of calculus): 

$$dZ = a(Z) dt + b(Z) \diamond dW,$$

with $Z(0) = \xi$ and $W := W_{\alpha, \eta, \beta}$ denotes a scalar Levy process.

We note that the topology used for the weak convergence in the above theorem is weaker than the $\mathcal{M}_1$ topology but is sufficiently strong to guarantee convergence in the sense of joint distributions. The theorem implies in particular convergence in distribution of $\hat{z}_\epsilon(t)$ to the stochastic process $Z(t)$ at fixed $t$. However, the theory does not provide convergence rates which would allow to better assess the required computational cost.

It was shown that the initial condition can be, in theory, equally well chosen using the invariant probability measure $\mu$ or the uniform Lebesgue measure. Empirically, it was found that convergence of the probability density at $t=1$ is faster if the initial conditions are drawn using $\mu$. Hence, following [Gottwald-Melbourne (2021)](https://epubs.siam.org/doi/10.1137/20M1333183), we work with the initial conditions drawn using $\mu$. The explicit formula for the density $\tilde{h}$ is less helpful due to the singular behavior near $x=0$. To circumvent this, we propagate uniformly distributed initial conditions $x_0' \in [0,1]$ under 10000 iterations of the Thaler map and then work with the initial conditions $x_0 = T^{10000} x_0'$. 

The following is an implementation of the above procedure.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import fsolve
from scipy.special import gamma as Gamma

def Thaler_map(x, gamma):
  return ((x**(1-gamma) + (1+x)**(1-gamma) - 1)**(1/(1-gamma))) % 1

def chi_func(x, x_star, p, seed):
  np.random.seed(seed)
  if x <= x_star:
    return 1.0
  else: 
    return 2*np.random.binomial(1,p,1)[0] - 1

def v_func(x, x_star, eta, gamma):
  alpha = 1/gamma
  if x <= x_star:
    return eta* ((1-2**(gamma-1))**(-gamma)) * ( (alpha**alpha)*(1-gamma)*Gamma(1-alpha)*np.cos(alpha*np.pi/2)/(2**(1-gamma)-1) )**(-gamma)
  else:
    return (eta* ((1-2**(gamma-1))**(-gamma)) * ( (alpha**alpha)*(1-gamma)*Gamma(1-alpha)*np.cos(alpha*np.pi/2)/(2**(1-gamma)-1) )**(-gamma))/(1-2**(1-gamma))

def obs_v(x_0, p, T, gamma, eta=1.0, seed=0):
  y = x_0
  def func(x, gamma=gamma):
   return x**(1-gamma) + (1+x)**(1-gamma) - 2
  x_star =  fsolve(func, (0.01))[0]
  chi_i = chi_func(x=x_0, x_star=x_star, p=p, seed=seed) 
  v_0 = chi_i*v_func(y, x_star=x_star, eta=eta, gamma=gamma)
  v_list = []
  v_list.append(v_0)

  for i in range(T):
    y = Thaler_map(y, gamma=gamma)
    chi_i = chi_i * chi_func(x=y, x_star=x_star, p=p, seed=seed)
    v_new = v_func(x = y, x_star=x_star, eta=eta, gamma=gamma)
    v = chi_i * v_new
    v_list.append(v)
  return v_list

#####################################
#choose gamma, beta, p, T, and seed
gamma = 0.625
beta=0.01
p=0.5*(1+beta)
T=100
seed=0
######################################

#choose IC following the paper
def init_data(gamma, seed=1):
  np.random.seed(seed)
  x = np.random.uniform()
  for i in range(10000):
    x = Thaler_map(x, gamma=gamma)
  x_0 = x
  return x_0

print('x_0: ', init_data(gamma=gamma, seed=seed))

#plot of the v^n
plt.rcParams["figure.figsize"] = (10,6)
v_list = obs_v(init_data(gamma=gamma, seed=seed), p=p, T=T, gamma=gamma, seed=seed)
print('sign of v_0: ', np.sign(v_list[0]))
plt.plot(v_list)
plt.title('observables v')
plt.show()

#iterating Thaler map itself  
thaler_iters = []
x = init_data(gamma=gamma, seed=seed)
thaler_iters.append(x)
for i in range(T):
  x = Thaler_map(x, gamma=gamma)
  thaler_iters.append(x)

plt.plot(thaler_iters)
plt.title('Thaler iterates x')
plt.show()

#simulate 
#sim1 = np.array([ (1/(100**(0.625)))*np.sum(obs_v(x_0=np.random.uniform(), x_star=x_star, p=p, T=100)) for i in range(50000)])
#plt.hist(sim1, bins = 100)
#plt.title('iter 1')
#plt.show()
```

## The widening valleys loss function 

Following [Orvieto et. al. (2022)](https://arxiv.org/abs/2202.02831v1), we will illustrate the dynamics of gradient descent perturbed by different perturbation schemes, including the one generated by deterministic dynamical systems as discussed earlier, on the widening valley loss landscape.

The loss is given by $L(u,v) = v^2 ||u||^2/2$. One can compute the gradient of $L$ to be given by $\nabla L(u,v) = (v^2 u, \|u\|^2 v)$, and the trace of the Hessian to be $d v^2 + \|u\|^2$, where $d$ is the dimension of $u$. The trace of the Hessian measures the flatness of the minima, which is monotonously changing in our case.  

The loss has a valley of minima with zero loss for all $(u,v)$ with $v=0$. The smaller the norm of $u$, the flatter the minimum. Gradient descent gets stuck when it first enters this valley. However, it was shown in the paper that injecting anticorrelated noise helps convergence to a flat minimum (with small $\|u\|$). All $(u,v)$ with $v=0$ are minima, but we also need $\|u\|$ to be minimized in order to get a small trace of the Hessian. In high dimension (large $d$), the gradient descent path is biased towards making $v$ small and not optimizing $u$ (the direction along $v$ is the most curved). 

The following illustrates the loss landscape:

```python
import matplotlib.pyplot as plt
from matplotlib import cm
plt.rcParams["figure.figsize"] = (10,6)

fig, ax = plt.subplots(subplot_kw={"projection": "3d"})

uline = np.linspace(0, 800, 1000)
vline = np.linspace(-10, 10, 1000)
u, v = np.meshgrid(uline, vline)
Loss = u*v**2/2
surf = ax.plot_surface(v, u, Loss, cmap=cm.coolwarm,
                       linewidth=0, antialiased=False, alpha=0.5)


fig.colorbar(surf, shrink=0.5, aspect=5)
ax.set_xlabel('v')
ax.set_ylabel('$\|\|u\|\|^2$')
ax.set_zlabel('L(u,v)')
plt.show()
```

## Experiments

**Setup.** Let $\{(x^{(i)}, y^{(i)})\}_{i=1}^N$ denote a dataset of $N$ input-output pairs with $x^{(i)} \in \mathbb{R}^{d_{in}}$ and $y^{(i)} \in \mathbb{R}$. We consider learning a neural network mapping $f_\theta : \mathbb{R}^{d_{in}} \to \mathbb{R}$, with parameters $\theta \in \mathbb{R}^p$ trained to minimize the empirical risk. 

Let $L^{(i)}: \mathbb{R}^p \to \mathbb{R}$ be the loss associated with the $i$th data point $(x^{(i)}, y^{(i)})$. We denote by $L(\theta) := \frac{1}{N} \sum_{i=1}^N L^{(i)}(\theta)$ the (full-batch) training loss, which we optimize using gradient descent to find the best parameters. 

We consider the perturbed gradient descent, which sequentially updates the parameters as:
$$\theta_{n+1} = \theta_n - \eta \nabla L(\theta_n) + \sigma y_{n+1},$$
where $\eta \geq 0$ is the learning rate, $\sigma > 0$ is the perturbation level, and $\{y_n\}_{n=0}^M$ are perturbations added to the iterates of vanilla gradient descent. Here $\theta := (u,v) \in \mathbb{R}^{d+1}$ for our case of widening valley loss. 

For the experiments, we start optimizing from the point $(u_0,0)$, where $u_0 \sim 5*Unif(d)$, and choose $d=10$ and learning rate $\eta = 0.01$. The goal is to test the effectiveness of different perturbation schemes in helping gradient descent to move to flat parts of the valley. 

We consider the following perturbation schemes: 
- baseline (vanilla gradient descent): $\sigma := 0$ 
- uncorrelated Gaussian noise injection: $\sigma > 0$ and the $y_n$ are i.i.d. standard Gaussians
- anticorrelated Gaussian noise injection: $\sigma > 0$ and the $y_{n+1} := \xi_{n+1} - \xi_n$, where the $\xi_n$ are i.i.d. standard Gaussians 
- simple deterministic perturbations: $\sigma > 0$ and the $y_{n} =  (-1)^n y_0$ for some $y_0 \neq 0$   
- perturbations generated by the Thaler map: $\sigma > 0$ and the $y_{n} = \eta^\gamma v^{(n)}$ with $\gamma = 1/\alpha \in (1/2, 1)$ defined earlier. The theorem (with $\epsilon := \eta$) then tells us that in the small learning rate limit the perturbed gradient descent converges in law to a gradient flow driven by $\alpha$-stable Levy process.

The following are the trajectories computed by perturbed gradient descent for an example of the widening valley loss under the above perturbation schemes after 20 iterations. Quantity of interest is the trace of the loss hessian (the smaller the better) after 20 iterations. We are also interested to see how fast the algorithm can move to a flat region.

```python
#baseline

import numpy as np
np.random.seed(0)

D = 5
eta = 0.01
d = 10
u = np.random.uniform(d)*D
v = 0.0

n_iter = 20
norm_list_0 = []
v_list_0 = []
trace_hessian_list_0 = []
loss_0 = []

for i in range(n_iter):
    print('Iter: ', i+1)
    u = u - eta * (u * v**2) 
    v = v - eta * (v * np.linalg.norm(u)**2) 
    trace_hessian_list_0.append((d-1)*v**2 + np.linalg.norm(u)**2)
    norm_list_0.append(np.linalg.norm(u)**2)
    v_list_0.append(v)
    loss_0.append((v**2)*np.linalg.norm(u)**2/2)

    print('2-norm squared of u: ', np.linalg.norm(u)**2)
    print('v: ', v)
    print('trace of loss Hessian: ', d*v**2 + np.linalg.norm(u)**2)

final_trace_baseline = d*v**2 + np.linalg.norm(u)**2 
```

```python
import matplotlib.pyplot as plt
from matplotlib import cm
plt.rcParams["figure.figsize"] = (10,6)

fig, ax = plt.subplots(subplot_kw={"projection": "3d"})

uline = np.linspace(0, 800, 1000)
vline = np.linspace(-10, 10, 1000)
u, v = np.meshgrid(uline, vline)
Loss = u*v**2/2
plt.plot(v_list_0, norm_list_0, loss_0, '-ro')
surf = ax.plot_surface(v, u, Loss, cmap=cm.coolwarm,
                       linewidth=0, antialiased=False, alpha=0.5)


fig.colorbar(surf, shrink=0.5, aspect=5)
ax.set_xlabel('v')
ax.set_ylabel('$\|\|u\|\|^2$')
ax.set_zlabel('L(u,v)')
plt.show()
```
